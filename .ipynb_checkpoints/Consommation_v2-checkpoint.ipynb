## 2.2 Regex
#### Create Regex match pattern

```python
valuelist=building_2015.Location.apply(str)
valuelist= ([s.replace("\"", "\'") for s in valuelist])

matchlist=[]
for regvalue in valuelist:
    matchlist.append(re.findall(pattern= ":\s\'([a-zA-Z0-9\s.-]+)" ,string=regvalue))
```

**create new feature from regex match**

```python
matchlist = np.array(matchlist)
building_2015_min["Latitude"]=matchlist[:,0].astype("float64")
building_2015_min["Longitude"]=matchlist[:,1].astype("float64")
building_2015_min["ZipCode"]=matchlist[:,5].astype("float64")    
```

#### Create a list of multiple model for the first test, to select the best model to optimise

```python
models = []
models.append(("LinearRegression",make_pipeline(preprocessor, LinearRegression())))
models.append(("LinearRegression Ridge",make_pipeline(preprocessor, Ridge())))
models.append(("LinearRegression Lasso",make_pipeline(preprocessor, linear_model.Lasso())))
models.append(("DecisionTree",make_pipeline(preprocessor, DecisionTreeRegressor(random_state=0))))
models.append(("RandomForest",make_pipeline(preprocessor, RandomForestRegressor(random_state=0))))
models.append(("XGBoost",make_pipeline(preprocessor, xgb.XGBRegressor())))
```

```python
names = []
results = [] #CV r2
results0 = [] #CV mape
results1 = [] #testset mape
results2= [] #testset r2

for name,model in models:
    
    #cross validation for 2 metrics (r2 and mape)
    kf=KFold(n_splits=3, random_state=0, shuffle=True)
    scores=cross_val_score(model, X_train, y_train1, cv=kf)
    scores_mape=cross_val_score(model, X_train, y_train1,cv=kf,scoring='neg_mean_absolute_percentage_error' )
    
    #add result to new list
    names.append(name)
    results.append(mean(scores))
    results0.append(mean(abs(scores_mape)))
    
    # fit and train to get test set result
    model_scv=model.fit(X_train,y_train1)
    y_pred=model_scv.predict(X_test)
    results1.append(mean_absolute_percentage_error(y_test1, y_pred))
    results2.append(r2_score(y_test1,y_pred ))

#create new dataframe with result 
d = {'model': names,'CV_r2':results,'testset_r2': results2,'CV_mape': results0 ,'testset_mape': results1}
result_t1=pd.DataFrame(data=d)
result_t1
```
